```python
import pandas as pd 
import plotly.express as px
```

```python
df = pd.read_csv('./cpdata.csv')
print(df.head(100))
```

```python
# Temperature Visualization
fig = px.bar(df, x='label', y='temperature', color='temperature')
fig.show()
```

```python
# Humidity visualization

fig2 = px.bar(df, x='label', y='humidity', color='humidity')
fig2.show()
```

```python
# ph visualization

fig3 = px.bar(df, x='label', y='ph', color='humidity')
fig3.show()
```
